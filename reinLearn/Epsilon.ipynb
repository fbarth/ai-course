```python
import pandas as pd

epsilon = 0.7
epsilon_dec = 0.99
epsilon_min = 0.1

df = pd.DataFrame(columns=['i','epsilon'])
i = 0
while epsilon > epsilon_min:
    df = df.append({'i': i, 'epsilon': epsilon}, ignore_index=True)
    epsilon = epsilon * epsilon_dec
    i = i + 1
```

```python
plt = df['epsilon'].plot(title='Valor de epsilon com decaimento de 0.99')
plt.set_xlabel('Iterações')
plt.set_ylabel('Epsilon')
```
